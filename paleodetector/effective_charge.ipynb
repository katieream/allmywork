```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Nov 12 16:18:55 2024

script to add in a column for effective charge per atom. Iterates through and identifies element by type number.
only set to work with our specific olivine set up, such that [1, 2, 3, 4] = [Si, Mg, Fe, O]

will add in oxidation states and the effect on the charge state later. Fe-O and O-O bonds in the LJ potential 
will be important for this!

the morse potentials will be affected by this, and the coulomb forces that are calculated via the charge
column in this input lattice will also be affected

if you change the total number of atoms, this needs to be manually changed in the .lmp AFTER modifying

@author: emlavoie
"""
```

```python
import re

with open('diamond_input.lmp', 'r') as fin, open("diamond_input_w_charge.lmp", "w") as fout:
    i = 0
    for line_no, line in enumerate(fin): 
        line_data = line.split()
        
        if len(line) < 16:
            string = line
            new_string = re.sub(' +', ' ', string)
            fout.write(new_string)
            
        elif line[14] == '1':
            print('Mg detected')
            i += 1
            string = line[:-64] + '    1.2' + line[-64:]
            new_string = re.sub(' +', ' ', string)
            
            new_string = " ".join(new_string.rsplit(None, -1)[1:])
            new_string = str(i) + ' ' + new_string+'\n'
            fout.write(new_string)
            
        elif line[14] == '2':
            print('Fe detected')
            i += 1
            string = line[:-64] + '    1.2' + line[-64:]
            new_string = re.sub(' +', ' ', string)
            new_string = " ".join(new_string.rsplit(None, -1)[1:])
            new_string = str(i) + ' ' + new_string +'\n'
            fout.write(new_string)
            
        elif line[14] == '3':
            print('O detected')
            i += 1
            string = line[:-64] + '   -1.2' + line[-64:]
            new_string = re.sub(' +', ' ', string)
            new_string = " ".join(new_string.rsplit(None, -1)[1:])
            new_string = str(i) + ' ' + new_string+'\n'
            fout.write(new_string)
            
        elif line[14] == '4':
            print('Si detected')
            i += 1
            string = line[:-64] + '    2.4' + line[-64:]
            new_string = re.sub(' +', ' ', string)
            new_string = " ".join(new_string.rsplit(None, -1)[1:])
            new_string = str(i) + ' ' + new_string+'\n'
            fout.write(new_string)
            
        else:
            print('no data line')
            string = line
            new_string = re.sub(' +', ' ', string)
            fout.write(new_string)
```
