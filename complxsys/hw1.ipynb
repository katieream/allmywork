```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
from random import random, randint
```

```python
#write the midpoint function which takes two tuples in as its input:
def midpoint(point1, point2):
    result = (0.5*(point1[0]+point2[0]), 0.5*(point1[1]+point2[1]))
    #print(f"The midpoint between {point1} and {point2} is {result}")
    return result

point1 = (0,0)
point2 = (1,1)
midpoint(point1, point2)

a, b, c = (0,0), (1,0), (0.5, np.sqrt(3/2))
corner = [a, b, c] #defining the points of the S triangle
x_coords, y_coords = zip(*corner)
N = 1000 # number of iterations to run the chaos game over
```

```python
#write a function to run the chaos game
def chaos(n_times):
    x, y = [], [] #initialize these as lists such that we can append to them
    in_point = (np.random.rand(), np.random.rand()) #generate starting random point within original triangle regime
    print("Initial point:", in_point)
    x.append(in_point[0]), y.append(in_point[1])
    
    #looping over number of times to play the chaos game
    for _ in range(n_times):
        choice = corner[np.random.choice(len(corner))]  # Randomly pick a corner
        newpoint = midpoint(in_point, choice)
        x.append(newpoint[0]), y.append(newpoint[1])
        in_point = newpoint
        
        
    plt.figure(figsize=(8, 8))
    plt.scatter(x_coords, y_coords, color = 'darkblue')
    plt.scatter(x, y, s=1.5, color='dodgerblue')
    plt.title(f"Chaos Game over {n_times} Iterations")
    plt.grid(True)
    print("\nThat was so much fun!")

N = 1000
chaos(N)
```
